### Calculate new columns

#### Ride Length

```python
df['started_at'] = pd.to_datetime(df['started_at'], format='mixed')
df['ended_at'] = pd.to_datetime(df['ended_at'], format='mixed')
df['ride_length'] = df['ended_at'] - df['started_at']
df.head(5)
```

```python
# Convert ride_length to hh:mm:ss format
df["ride_length"] = df["ride_length"].apply(
    lambda x: f"{x.components.hours:02}:"
              f"{x.components.minutes:02}:"
              f"{x.components.seconds:02}"
)
df.drop(columns=['hhmmss'], inplace=True)
df.head(5)
```

#### Day of week

```python
# calculate the day of the week for each ride started_at
# df['day_of_week'] = df['started_at'].dt.day_name()

# encode the day of the week such that Monday=1, Tuesday=2, ..., Sunday=7
df['day_of_week'] = df['started_at'].dt.weekday + 1
df.head(5)
```
